```python
import cv2
import pytesseract

# Đọc ảnh
image = cv2.imread('data/1/img/20140603_0003_BCCTC_tg_0_1.png')

# Chuyển ảnh sang màu xám
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Áp dụng nhị phân hóa để làm nổi bật văn bản
_, binary = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
plt.imshow(image)
```

```python
import matplotlib.pyplot as plt
```

```python
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

word_images = []
for contour in contours:
    # Tính toán bounding box cho mỗi contour
    x, y, w, h = cv2.boundingRect(contour)
    print(x, y, w, h)
    # Cắt từng từ
    word_image = image[y:y+h, x:x+w]
    word_images.append(word_image)

    # Hiển thị từ vừa cắt
    plt.imshow(word_image)
    plt.show()
```
